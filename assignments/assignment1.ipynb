**Question 4:** Write a Python program to find the largest of three numbers.

```python
a=float(input("Enter first number:")) #user input:1
b=float(input("Enter second number:")) #user input:2
c=float(input("Enter third number:")) #user input:3
lar=0
if a>b and a>c:
    lar=a
if b>a and b>c:
    lar=b
if c>b and c>a:
    lar=c
print("The largest amoung the three numbers given is:",lar)
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
n=int(input("Enter any positive number:")) #user input:4
a=1
for i in range(1,n+1):
    a=a*i
print("The factorial of the given number is:",a)
```
