# Simple Calculator


```python
num1= float(input("Enter the first number:"))
num2= float(input("Enter the second number:"))
add=num1+num2
sub=num1-num2
multiply=num1*num2
divide=num1/num2
floor_div=num1//num2
mod=num1%num2
print("Addition:",add)
print("Substraction:",sub)
print("Multiplication:",multiply)
print("Division:",divide)
print("Floor division:",floor_div)
print("Reminder:",mod)
```

# USING CONDITIONAL STATEMENTS



```python
num1= float(input("Enter the first number:"))
num2= float(input("Enter the second number:"))
op=input("enter any one of the operators: + - / % // ")
if op=="+":
    ans=num1+num2
    print(ans)
elif op=="-":
    ans=num1-num2
    print(ans)
elif op=="*":
    ans=num1*num2
    print(ans)
elif op=="/":
    if num2!=0:
        ans=num1/num2
        print(ans)
    else:
        print("cannot be divided with zero")
elif op=="//":
    ans=num1//num2
    print(ans)
elif op=="%":
    ans=num1%num2
    print(ans)
else:
    print("enter proper operator")
```
